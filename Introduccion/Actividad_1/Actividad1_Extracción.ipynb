## 1️⃣ JSON: Información de empleados

Supón que tienes el archivo <code>empleados.json</code> con la siguiente estructura:

```json
[
  {"nombre": "Ana", "edad": 28, "departamento": "Ventas", "salario": 2500},
  {"nombre": "Luis", "edad": 35, "departamento": "IT", "salario": 3200},
  {"nombre": "María", "edad": 24, "departamento": "Ventas", "salario": 2300},
  {"nombre": "Carlos", "edad": 40, "departamento": "IT", "salario": 4000}
]
```

<b>Ejercicios:</b>

1. Crea un DataFrame a partir de la extracción de información de este archivo JSON.
2. ¿Cuál es el <b>salario promedio por departamento</b>?
3. ¿Quién es el <b>empleado más joven</b> y a qué departamento pertenece?
4. ¿Cuántos empleados tienen un <b>salario mayor a 2500</b>?
5. ¿Cuál es la <b>diferencia de edad</b> entre el empleado más joven y el de mayor edad?
6. ¿Cuál es el <b>salario total pagado</b> en la empresa?

<i>Tip: Usa pandas.read_json o json.load para cargar los datos.</i>

````python
# Tu código aquí
import pandas as pd
````

````python
#1 DataFrame
df_json = pd.read_json("empleados.json") # Leer el archivo JSON 
print("DataFrame de EMPLEADOS:")
print(df_json)
````

````python
#2 Salario promedio por departamento
salario_promedio = df_json.groupby("departamento")["salario"].mean() #Agrupar por departamento y sacar promedio
print("\nSalario promedio por departamento:")
print(salario_promedio)
````

````python
#3 Empleado más joven y a qué departamento pertenece
empleado_joven = df_json.loc[df_json["edad"].idxmin()] 
print(f"\nEmpleado más joven: {empleado_joven['nombre']}, Departamento: {empleado_joven['departamento']}")
````

````python
#4 empleados con salario mayor a 2500
mayores_2500 = df_json[df_json["salario"] > 2500].shape[0]
print(f"\nEmpleados con salario mayor a 2500: {mayores_2500}")
````

````python
#5 Diferencia de edad entre el empleado más joven y el de mayor edad
diferencia_edad = df_json["edad"].max() - df_json["edad"].min() #Edad mayor- Edad menor
print(f"\nDiferencia de edad entre el empleado más joven y el de mayor edad: {diferencia_edad} años")
````

````python
#6 Salario total pagado en la empresa
salario_total = df_json["salario"].sum() #Suma salarios
print(f"\nSalario total pagado en la empresa: {salario_total}")
````

## 2️⃣ XML: Catálogo de libros

Supón que tienes el archivo <code>libros.xml</code> con la siguiente estructura:

```xml
<libros>
  <libro>
    <titulo>Python para Todos</titulo>
    <autor>Juan Pérez</autor>
    <precio>29.99</precio>
    <anio>2020</anio>
  </libro>
  <libro>
    <titulo>Data Science Avanzado</titulo>
    <autor>María López</autor>
    <precio>45.50</precio>
    <anio>2022</anio>
  </libro>
  <libro>
    <titulo>Ingeniería de Características</titulo>
    <autor>Pedro Ruiz</autor>
    <precio>39.00</precio>
    <anio>2019</anio>
  </libro>
</libros>
```

<b>Ejercicios:</b>

1. Crea un DataFrame a partir de la extracción de información de este archivo XML.
2. ¿Cuál es el <b>libro más caro</b> y su autor?
3. ¿Cuál es el <b>precio promedio</b> de los libros publicados después de 2020?
4. ¿Cuántos libros fueron <b>publicados antes de 2021</b>?
5. ¿Cuál es la <b>diferencia de precio</b> entre el libro más barato y el más caro?
6. ¿Cuántos <b>autores diferentes</b> hay en la lista?

<i>Tip: Usa xml.etree.ElementTree para leer el archivo y pandas para tabular los datos.</i>

````python
# Tu código aquí
import xml.etree.ElementTree as ET #biblioteca para leer, analizar, crear y modificar archivos XML.
````

````python
tree = ET.parse('libros.xml')  # Abre y analiza el archivo XML
root = tree.getroot()          # Obtiene el elemento raíz del documento

registros= [] #Diccionario para guardar los libros
print('LIBROS:')
for libro in root.findall('libro'): #Leer libro a libro
    titulo = libro.find('titulo').text
    autor = libro.find('autor').text
    precio = float(libro.find('precio').text)
    anio = int(libro.find('anio').text)
    print(f'Título: {titulo}, Autor: {autor}, Precio: {precio}, Año: {anio}') 
    #Guardar la información
    registros.append({ 
        'Título': titulo,
        'Autor': autor,
        'Precio': precio,
        'Año': anio
    })
````

````python
#1 Crea un DataFrame a partir de la extracción de información de este archivo XML
df_libros = pd.DataFrame(registros)
print("DataFrame de libros:\n", df_libros, "\n")
````

````python
#2 Libro más caro y su autor?
libro_caro = df_libros.loc[df_libros['Precio'].idxmax()] #Buscar el libro por posicion de precio mas alto (idmax)
#.loc fila completa
print("Libro más caro:", libro_caro['Título'], "- Autor:", libro_caro['Autor'])
````

````python
#3 Precio promedio de los libros publicados después de 2020
publicados_2020 = df_libros[df_libros['Año'] > 2020]['Precio'].mean() #Filtra libros publicados después del año 2020 y saca promedio del precio
print("Precio promedio de libros después de 2020:", publicados_2020)
````

````python
#4 Libros publicados antes de 2021
libros_2021 = df_libros[df_libros['Año'] < 2021].shape[0] #Filtra libros con año menor a 2021 .shape(numero de filas)
print("Libros publicados antes de 2021:", libros_2021)
````

````python
#5 Diferencia de precio entre el libro más barato y el más caro
diferencia_precio = df_libros['Precio'].max() - df_libros['Precio'].min() #Precio mas alto-Precio mas bajo
print("Diferencia de precio entre el más barato y el más caro:", diferencia_precio)
````

````python
#6 Autores diferentes 
autores_diferentes = df_libros['Autor'].nunique() #Columna de autores y cuenta los distintos
print("Autores diferentes:", autores_diferentes)
````

## 3️⃣ CSV: Ventas detalladas

Supón que tienes el archivo <code>ventas_detalladas.csv</code> con la siguiente estructura:

```csv
producto,fecha,cantidad,precio_unitario
A,2023-01-01,10,2.5
B,2023-01-02,5,3.0
A,2023-01-03,7,2.5
C,2023-01-01,3,4.0
B,2023-01-03,2,3.0
```

<b>Ejercicios:</b>

1. Crea un DataFrame con los datos extraídos del archivo.
2. ¿Cuál es el <b>total vendido</b> por producto y por fecha?
3. ¿En qué fecha se vendió la <b>mayor cantidad de productos</b>?
4. ¿Cuál es el <b>producto con mayor ingreso total</b>?
5. ¿Cuál es la <b>media de cantidad vendida</b> por producto?
6. ¿Qué producto tuvo <b>ventas todos los días registrados</b>?

<i>Tip: Usa pandas.read_csv y agrupa los datos según lo que se pide.</i>

````python
import pandas as pd #Libreria
````

````python
# 1 DataFrame con los datos extraídos del archivo.

df_ventas = pd.read_csv('ventas_detalladas.csv') #Leer el archivo CSV
print("DataFrame original:\n", df_ventas, "\n")
````

````python
#2 Total vendido por producto y por fecha

#Agrupar los datos por producto y fecha, selecciona cantidad-precio unitario (suma)
t_prod_fecha = df_ventas.groupby(['producto', 'fecha'])[['cantidad', 'precio_unitario']].sum() 
print("Total por producto y fecha:\n", t_prod_fecha, "\n")
````

````python
#3 Fecha donde se vendió la mayor cantidad de productos

#Sumar todas las cantidades vendidas por fecha
fecha_mayor_venta = df_ventas.groupby('fecha')['cantidad'].sum().idxmax() #idmax (valor mayor)
print("Fecha con mayor cantidad de productos vendidos:", fecha_mayor_venta)
````

````python
#4 Producto con mayor ingreso total

df_ventas['ingreso'] = df_ventas['cantidad'] * df_ventas['precio_unitario'] #Nueva columna ingreso calculando cantidad * precio_unitario
producto_mayor_ingreso = df_ventas.groupby('producto')['ingreso'].sum().idxmax() #suma todos los ingresos por producto .idxmax(mayor ingreso)

print("Producto con mayor ingreso total:", producto_mayor_ingreso)
````

````python
#5 Media de cantidad vendida por producto

#Agrupar por producto y calcula el promedio de cantidad vendida.
media_producto = df_ventas.groupby('producto')['cantidad'].mean() #mean (promedio)
print("Media de cantidad vendida por producto:\n", media_producto, "\n")
````

````python
#6 Producto con ventas todos los días registrados

# Número total de fechas diferentes
t_fechas = df_ventas['fecha'].nunique()
print ("Fechas diferentes",t_fechas)

# Conteo de fechas en las que participa cada producto
p_ventas_dias = df_ventas.groupby('producto')['fecha'].nunique()
p_ventas_dias = p_ventas_dias[
p_ventas_dias == t_fechas
]

print("\nProducto con ventas todos los días registrados:\n", p_ventas_dias)

#No muestra información porque ningún producto se vendió todos los días registrados
````

## 4️⃣ Imágenes: Análisis de pixeles

Supón que tienes la imagen <code>alinco.png</code> en la misma carpeta que este notebook.

<b>Ejercicios:</b>

1. Extrae la información de los pixeles de la imagen y muestra sus dimensiones.
2. ¿Cuál es el <b>valor RGB</b> del píxel (50, 50)?
3. ¿Cuál es el <b>color predominante</b> en la imagen?

<i>Tip: Usa matplotlib.pyplot.imread o cv2.imread para cargar la imagen y numpy para analizar los datos.</i>

````python
#Librerias
import numpy as np #Matrices y operaciones numéricas
import matplotlib.pyplot as plt #Para mostrar gráficos o imágenes
from matplotlib.image import imread #Carga imágenes en forma de matriz numpy
````

````python
#Cargar imagen
#Carga la imagen como matriz numpy y se guarda en la variable imagen
imagen = imread('alinco.png') 
````

````python
#1 Extrae la información de los pixeles de la imagen y muestra sus dimensiones.

# imagen.shape (alto, ancho, canales) dimensiones de la imagen
alto, ancho, canales = imagen.shape
print("Dimensiones de la imagen:", alto, "x", ancho, "pixeles")
print("Número de canales:", canales)  # 3 → RGB, 4 → RGBA
````

````python
#2 Valor RGB del píxel (50, 50)

#imagen[fila, columna] accede al color de un píxel específico.
pixel_50_50 = imagen[50, 50]  # filas, columnas (Fila 50, columna 50 píxel en esa posición.)
print("Valor RGB del píxel (50,50):", pixel_50_50) #pixel_50_50 será un array con los valores de los canales
````

````python
#3 Color predominante en la imagen

# Aplanar la imagen a lista de píxeles
pixeles = imagen.reshape(-1, canales)  # cada fila = un pixel

colores, counts = np.unique(pixeles, axis=0, return_counts=True) # Contar colores únicos
color_predominante = colores[np.argmax(counts)]
print("Color predominante en la imagen (RGB):", color_predominante)
````
